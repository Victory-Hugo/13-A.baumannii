```python
import pandas as pd
from Bio import SeqIO
```

```python
BASENAME = "ERR1946991"
PROKKA_DIR = "/mnt/d/1-鲍曼菌/注释prokka/"
ORIGIN_FILE = f"{PROKKA_DIR}/{BASENAME}/{BASENAME}.ffn"
```

```python
# 首先读取原始的ffn文件
# 获取ffn文件中的每一个序列的ID
seq_ids = []
for record in SeqIO.parse(ORIGIN_FILE, "fasta"):
    seq_ids.append(record.id)
seq_ids
# 将BASENAME添加到序列ID的前面
seq_ids = [f"{BASENAME}|{seq_id}" for seq_id in seq_ids]
seq_ids
```

```python
# 现在读取输出的K_locus的fna结果
# 并获取序列的ID
K_DIR = "/mnt/d/1-鲍曼菌/荚膜多糖/K_locus_results"
K_FILE = f"{K_DIR}/{BASENAME}_kaptive_results.fna"
K_seq_ids = []
for record in SeqIO.parse(K_FILE, "fasta"):
    K_seq_ids.append(record.id)
K_seq_ids
# 去掉:及之后的内容
K_seq_ids = [seq_id.split(":")[0] for seq_id in K_seq_ids]
K_seq_ids
```

```python
# 现在读取输出的OCL_locus的fna结果
# 并获取序列的ID
OCL_DIR = "/mnt/d/1-鲍曼菌/荚膜多糖/OCL_results"
OCL_FILE = f"{OCL_DIR}/{BASENAME}_kaptive_results.fna"
OCL_seq_ids = []
for record in SeqIO.parse(OCL_FILE, "fasta"):
    OCL_seq_ids.append(record.id)
OCL_seq_ids
# 去掉:及之后的内容
OCL_seq_ids = [seq_id.split(":")[0] for seq_id in OCL_seq_ids]
OCL_seq_ids
```

```python
# 先求OCL_seq_ids和K_seq_ids的并集，命名为OCL_K_union_ids
OCL_K_union_ids = set(OCL_seq_ids) | set(K_seq_ids)
OCL_K_union_ids
# 再求OCL_K_union_ids和seq_ids的差集，命名为OCL_K_diff_ids
OCL_K_diff_ids = set(seq_ids) - OCL_K_union_ids  
OCL_K_diff_ids
# 这里的OCL_K_diff_ids即为排除了OCL和K的序列ID
OCL_K_diff_ids
# 将OCL_K_diff_ids的|之前的内容去掉
OCL_K_diff_ids = [seq_id.split("|")[1] for seq_id in OCL_K_diff_ids]
OCL_K_diff_ids
# 根据OCL_K_diff_ids提取出ORIGIN_FILE中对应的序列到
OCL_K_diff_seqs = [record for record in SeqIO.parse(ORIGIN_FILE, "fasta") if record.id.split("|")[0] in OCL_K_diff_ids]
# 另存为 /mnt/d/1-鲍曼菌/毒力因子其他/BASENAME.Remain.ffn
SeqIO.write(OCL_K_diff_seqs, f"/mnt/d/1-鲍曼菌/毒力因子其他/{BASENAME}.Remain.ffn", "fasta")
```
